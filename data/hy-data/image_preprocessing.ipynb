```python
import cv2
import glob
import os
import shutil
from tqdm import tqdm
import numpy as np
```

```python
'''
Transfer stereo camera images from 1 folder to 2 foldersm based on the camera id
'''

read_path = "all_images/cam1-color/*"
cam1_path = "cam1"
cam2_path = "cam2"

i=0
for img_path in glob.glob(read_path):
    camera_id = int(img_path[-5])
    img_name = os.path.split(img_path)[1]
    if camera_id==1:
        new_path = os.path.join(cam1_path, img_name)
        shutil.copyfile(img_path, new_path)
    if camera_id==2:
        new_path = os.path.join(cam2_path, img_name)
        shutil.copyfile(img_path, new_path)
    i = i + 1

    if (i%100)==0:
        print("Transferred:", i)
```

```python
'''
Visualize images as a video
'''

# Video parameters
fourcc = cv2.VideoWriter_fourcc(*'mp4v') 
frameSize = (848, 800)
video = cv2.VideoWriter('cam1_video.avi', cv2.VideoWriter_fourcc(*'DIVX'), 50, frameSize)

# Read and sort images
images_list = glob.glob("cam1/*png")
images_list.sort()

# Read images and create video
for image_path in images_list:
  img = cv2.imread(image_path)
  img_resized = cv2.resize(img, frameSize, interpolation = cv2.INTER_AREA)
  video.write(img_resized)

video.release()
```
